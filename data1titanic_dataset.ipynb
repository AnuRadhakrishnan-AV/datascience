```python
x=data[data['Pclass']==1].Sex.value_counts()
x
```

```python
# Find out how many female passengers had travelled in first class and show on pie chardata
x=data[data['Pclass']==1].Sex.value_counts()
plt.pie(x)
plt.title("Pclass passengers")
plt.legend(["male","female"])
plt.show()
```
