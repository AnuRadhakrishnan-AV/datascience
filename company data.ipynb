```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv("company_sales_data.csv")
```

```python
# read total profit of all months and show it in a line graph 
#line style dotted clr green shoew legend at lower right add a title add cycle marker line width 
# should be 3,change marker, color ,style marker size
x=data['month_number']
y=data['total_profit']
plt.plot(x,y,color='green',linestyle='dotted',linewidth=3,marker='d',markerfacecolor='red',ms=8)
plt.xlabel('month number')
plt.ylabel('total profit')
plt.title('Month Profit Graph')
plt.legend('profit',loc='lower right')
plt.xticks(x,['jan','feb','mar','aprl','may','jun','jul','aug','sep','oct','nov','dec'])
plt.show()
```

```python
# read all the product sales data and show in a multiline plot
x=data['month_number']
y1=data['facecream']
y2=data['facewash']
y3=data['toothpaste']
y4=data['bathingsoap']
y5=data['shampoo']
y6=data['moisturizer']
data.plot(x='month_number',y=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])
plt.show()
```

```python
# read toothpaste sales data of each month and show it using scatter plot with grid lines
x=data['month_number']
y=data['toothpaste']
plt.scatter(x,y)
plt.grid(True)
plt.xticks(x,['jan','feb','mar','aprl','may','jun','jul','aug','sep','oct','nov','dec'])
plt.show()
```

```python
# read facecream and facewash product in a bar graph ,do usind data and plt data is a simple method instead of plt in data we can use data.
x=data['month_number']
y=data[['facecream','facewash']]
data.plot(x='month_number',y=['facecream','facewash'],kind='bar',color=['pink','black'],grid=True)
plt.show()
```

```python
# read the total profit of each month and show it using a histogram and most common profit range
x=data['total_profit']
y=[100000,200000,300000,400000,500000]
plt.hist(x,y)
plt.show()
```
