```python
x=[10,9,8,2,3,23,1,7,1,2]
y=[100,110,120,130,140,108,104,100,140,120]
m,c,r,p,std_err=stats.linregress(x,y)
def fun(x):
    return (m*x)+c
p=list(map(fun,x))
plt.scatter(x,y)
plt.plot(x,p)
plt.show()
```
