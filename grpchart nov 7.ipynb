```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([2,3,4,7,8])
y1=np.array([1,9,8,3,2])
y2=np.array([9,8,2,4,1])
plt.bar(x-0.2,y1,width=0.4)
plt.bar(x+0.2,y2,width=0.4)
plt.show()
```

```python
x=np.array([2,3,4,7,8])
y1=np.array([1,9,8,3,2])
y2=np.array([9,8,2,4,1])
y3=np.array([3,9,7,8,4])
plt.bar(x-0.2,y1,width=0.4,color="pink")
plt.bar(x+0.2,y2,width=0.4,color="black")
plt.legend(["a","b"])
plt.xlabel("Duration")
plt.ylabel("temp")
plt.title("BAR GRAPH")
plt.xticks(x,["a","b","c","d","e"])
plt.show()
```

```python
x=np.array(['a','b','c','d','e'])
y1=np.array([1,9,8,3,2])
y2=np.array([9,8,2,4,1])
plt.bar(x,y1)
plt.bar(x,y2,bottom=y1,color="pink")
plt.legend(["a","b"])
plt.xlabel("Duration")
plt.ylabel("temp")
plt.title("BAR GRAPH")
plt.show()
```

```python
x=np.array(['a','b','c','d','e'])
y1=np.array([10,9,8,3,2])
y2=np.array([30,8,2,4,1])
y3=np.array([48,8,4,9,7])
y4=np.array([9,1,2,8,3])
plt.bar(x,y1)
plt.bar(x,y2,bottom=y1)
plt.bar(x,y3,bottom=y1+y2)
plt.bar(x,y3,bottom=y1+y2+y3)

plt.show()
```
