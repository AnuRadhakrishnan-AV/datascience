```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot1

x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)

# plot2

x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
# plot1

x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,3,1)
plt.plot(x,y)
plt.title('FIRST')

# plot2

x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,3,2)
plt.plot(x,y)
plt.title('SECOND')

# plot3
x=np.array([0,1,2,3])
y=np.array([10,2,3,4])
plt.subplot(2,3,3)
plt.plot(x,y)
plt.title('THIRD')


# plot4
x=np.array([0,1,2,3])
y=np.array([10,2,3,8])
plt.subplot(2,3,4)
plt.plot(x,y)
plt.title('FOURTH')


# plot1+4
x=np.array([0,1,2,3])
y=np.array([10,9,2,8])
plt.subplot(2,3,5)
plt.plot(x,y)
plt.title('FIFTH')


# plot2+4
x=np.array([0,1,2,3])
y=np.array([10,9,2,8])
plt.subplot(2,3,6)
plt.plot(x,y)
plt.title('SIXTH')
plt.suptitle('PLOTTING')
plt.show()

```
